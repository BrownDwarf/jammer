$R\equiv \lambda/\delta \lambda$  
$\delta \lambda = \lambda/R$

```python
res_grad_dat['d_lam_300mas'] = res_grad_dat.wavelength_um / res_grad_dat.resolution_R
```

```python
plt.plot(res_grad_dat.wavelength_um, res_grad_dat.d_lam_300mas)
plt.xlabel('$\lambda \;(\mu\mathrm{m})$ ')
plt.ylabel('$\delta \lambda \;(\mu\mathrm{m})$ ')
```
